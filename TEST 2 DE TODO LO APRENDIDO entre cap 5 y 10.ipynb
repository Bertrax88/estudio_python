# 1. CONTROL DE FLUJO

1.1 Genere un dataframe de 15 nombres  y sus generos, con notas aleatorias siguiendo una normal con media en 5 y desviacion estandar 1, esto para notas de castellano, ciencias y matematicas

1.2 Entregue un dataframe solo con aquellos que tienen sobre 6.1 en matematicas (usando for e if)

1.3 Genere una nueva columna donde salga el promedio general de todos, matematicas pondera 50%, ciencias 20% y castellano 30%

1.4 recorra con for el dataframe y ponga "aprobado" o "reprobrado" si el promedio es menor a 4.5

1.5 ordene de mejor nota a peor nota promedio final

```python
#1.1
import pandas as pd 
import numpy as np 
nombres=['Diego','Ruben','Maria Jose','Daniel','Damian','Constanza','Hernan','Javiera','Francisca','Sergio','Macarena','Roberto','Andrea','Maria Paz','Jose']
apellidos=['Bertin','Bertin','Silva','Bertin','Bertin','Ljubetic','Olmedo','Gonzalez','Molina','Osorio','Osorio','Zahr','Gonzalez','Lagos','Vergara']
genero=['M','M','F','M','M','F','M','F','F','M','F','M','F','F','M']
DICC_NOTAS={'NOMBRES':nombres,'APELLIDOS':apellidos,'GENERO':genero}
DF_NOTAS=pd.DataFrame(DICC_NOTAS)
DF_NOTAS['CASTELLANO']=np.round(np.clip(np.random.normal(5.5,2,len(DF_NOTAS)),1,7), decimals=1)
DF_NOTAS['MATEMATICAS']=np.round(np.clip(np.random.normal(6,1.1,len(DF_NOTAS)),1,7), decimals=1)
DF_NOTAS['CIENCIAS']=np.round(np.clip(np.random.normal(5,2,len(DF_NOTAS)),1,7), decimals=1)
DF_NOTAS
```

```python
DF_NOTAS.iloc[3:4,:]
```

```python
DF_NOTAS.iloc[3,:]
```

```python
#1.2
df_MATEMATICAS = pd.DataFrame()
for i in range(len(DF_NOTAS)):
    if DF_NOTAS.iloc[i,4]>=6.1:   
        df_MATEMATICAS = pd.concat([df_MATEMATICAS, DF_NOTAS.iloc[i:i+1,:]])
df_MATEMATICAS.reset_index(drop=True, inplace=True)
df_MATEMATICAS
```

```python
# 1.3 nueva columna

for i in range(len(DF_NOTAS)):  
    DF_NOTAS.loc[i,'PROMEDIO']=0.5*DF_NOTAS.loc[i,'MATEMATICAS']+0.2*DF_NOTAS.loc[i,'CIENCIAS']+0.3*DF_NOTAS.loc[i,'CASTELLANO']
DF_NOTAS
```

```python
#1.4
for i in range(len(DF_NOTAS)):  
    if DF_NOTAS.loc[i,'PROMEDIO']>=4.5:
        DF_NOTAS.loc[i,'STATUS']='APROBADO'
    else:
        DF_NOTAS.loc[i,'STATUS']='REPROBADO'
DF_NOTAS
```

```python
#1.5
DF_NOTAS.sort_values(by= 'PROMEDIO',ascending=False)
DF_NOTAS.reset_index()
```

```python

# bonus, crear una columna con notas de musica, los primeros 7 del index con media 5 y los otros 8 con media 3
j=6
for i in range(len(DF_NOTAS)):
    if i<=j: 
        DF_NOTAS.loc[i,'MUSICA']=np.round(np.clip(np.random.normal(6,1,1),1,7), decimals=1)
    else: 
        DF_NOTAS.loc[i,'MUSICA']=np.round(np.clip(np.random.normal(3,2,1),1,7), decimals=1)
DF_NOTAS
```
